## Importing Libraries and Loading Data

```python
# Data manipulation and analysis library
import pandas as pd
# Linear algebra library
import numpy as np
# Create visualizations
import matplotlib as mpl
# Plotting
import matplotlib.pyplot as plt
# Inline plots
%matplotlib inline
# Statistical graphics
import seaborn as sns
# Visualizing missing data
import missingno as msno

np.set_printoptions(precision=2)
```

```python
train = pd.read_csv('Stores.csv')
```

We have $n = 5$ features, and $m = 895$ examples. We'll drop the `store_id` column because it has a non-predictive nature (does not correlate with the target variable).

```python
# The first column as a index object
first_column = train.columns[[0]]
first_column
# Remove the first column along the columns on the existing data frame
train.drop(first_column, axis = 1, inplace=True)
```

What is left now is the following features:

- Store_Area: Physical area of the store in yard
- Items_Available: Number of items available in the store
- Daily_Customer_Count: Number of customers who visit the store, on a month average
- Store_Sales: Sales in $ that the store made

## Exploratory Data Analysis
In this step, we'll examine and analyze the dataset to summarize its main characteristics, using visual methods and statistics. The goal of EDA is to understand the data, identify patterns and anomalies, and formulate hypothesis for further analytics or predictive modeling.

We'll start by considering some statistical data, using `pd.describe()`.

```python
train.describe()
```

**Hypothesis**: There are outliers in the form of very small and very large supermarkets, are they really outliers or just seperate groups of supermarkets?

We'll checks for gaps in the data using the `msno` library.

```python
msno.bar(train, figsize=(16,5), color='orange')
plt.show()
```

The diagram show that there are **no gaps** in the data.

Let us also check for categorical features, these are features that are non-numbers, and typically have a limited, fixed number of possible values (categories).

```python
for column_name in train.columns:
    unique_values = len(train[column_name].unique())
    print(f'Feature {column_name} has {unique_values} unique values.')
```

We deduce there are no categorical features. Let us analyze the statistical distribution of features.

```python
for column_name in train.columns:
    plt.figure(figsize=(7,7))
    plt.title(f"{column_name} distribution")
    # Create a density histogram 
    sns.histplot(train[column_name], stat='density', color='green')
    # Create a KDE plot
    sns.kdeplot(train[column_name], color='black')
    # Create a vertical line for the mean
    plt.axvline(train[column_name].mean(), color='red', linestyle='--', linewidth=0.8)
    # Min y value, max y value
    min_ylim, max_ylim = plt.ylim()
    # Create a textbox for the mean
    plt.text(train[column_name].mean() * 1.05, max_ylim * 0.96, f'Mean: {train[column_name].mean():.2f}')

    plt.xlabel('score')
    plt.ylabel('density')
    plt.show()
```

Normal distribution, this is good for linear regression.

We'll now examine any possible correlations between the features.

```python
# Set some default styles
sns.set()
# Create plot for each pair features in the dataframe, add a linear regression line, and for diagonal elements use KDE
sns.pairplot(train, height=5, kind='reg', diag_kind='kde')
plt.show()
```

**Conclusions**
- There is a strong correlation, between an increase in `store_area`, and and increase in `items_available`
- There is a slight correlation between `store_area` (`items_available`) and `store_sales`
- The `daily_customer_count` and `store_area` are not related

We'll analyze the presence of outliers in the data, according to the unspoken rule, we should not delete more than 2% of the training examples (no more than 18 examples in our case). 

```python
plt.figure(figsize=(15,25))
for i, feature in enumerate(train.columns):
    plt.subplot(6,3,i + 1)
    sns.boxplot(y=train[feature], color='green')
```

We can carry out a correlation analysis of features.

```python
correlation = train.corr()
print(correlation['Store_Sales'].sort_values(ascending=False))
```

```python
k = 10
# Get index of kth most correlated features with `Store_Sales`
cols = correlation.nlargest(k, 'Store_Sales')['Store_Sales'].index
print(cols)
# Calculate correlation between rows as numpy array
cm = np.corrcoef(train[cols].values.T)
fig, ax = plt.subplots(figsize=(14,12))
# Heatmap to represent `cm`
sns.heatmap(cm, vmax=.8, linewidths=0.01, square=True, annot=True, cmap='viridis', linecolor='white',
            xticklabels=cols.values, annot_kws={'size':12}, yticklabels=cols.values
           )
```

There is a very strong correlation between `Store_Area` and `Items_Available`. We should remove one of them, well choose `Store_Area`.

```python
train = train.drop('Store_Area', axis=1)
train
```

The training set consists of $n=3$ features now.

## Cluster Analysis

```python
from sklearn.cluster import KMeans
wcss = []
for k in range(1,11):
    km = KMeans(n_clusters=k, n_init='auto')
    km.fit_predict(train)
    wcss.append(km.inertia_)

fig, ax = plt.subplots()
ax.plot(range(1,11), wcss)
plt.show()
```

From the elbow method, we choose `k=4`.

## Cleaning the data from outliers
